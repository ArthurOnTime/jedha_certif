# Get meteo by cities using an API

```python
# Get the cities lat & lon from the Nominatim API
cities_json = []
# Iterate over the list of cities
for city in cities:
    # Query the API
    city_res = requests.get('https://nominatim.openstreetmap.org/search?city={}&format=json&country=france'.format(city))
    city_json = city_res.json()
    cities_json.append(city_json[0])
    # Check request success
    print("{} - {}".format(city, city_res))
    # Chill
    time.sleep(3)
# Create a DataFrame with lat & lon for all the cities
cities_df = pd.DataFrame(cities_json)
cities_df
```

```python
# Query the meteo API to get the cities precipitation prevision

# API Key >> Should be in .env
# ------------------------ WARNING KEY -------------------------
appid='***'
# --------------------------------------------------------------

# Iterate over the cities DataFrame to add meteo data
for i in range(len(cities_df)):
    # Get the row lat & lon required to query the meteo API for a city
    lat = cities_df.loc[i,'lat']
    lon = cities_df.loc[i,'lon']
    # Query the meteo API and convert to JSON for a city
    city_meteo_res = requests.get("https://api.openweathermap.org/data/2.5/forecast?lat={}&lon={}&appid={}".format(lat,lon,appid))
    city_meteo_json= city_meteo_res.json()
    # Extract average Proba Of Precipitation (pop) for the city from the meteo JSON
    sum_pop = 0
    for item in city_meteo_json["list"]: sum_pop += int(item['pop'])
    avg_pop = sum_pop / len(city_meteo_json["list"])
    # Append the city pop to the city's DataFrame row
    cities_df.loc[i,'meteo'] = avg_pop
    # Chill to keep the API unstressed
    time.sleep(3)

cities_df
```

```python
# Sorting cities by least POP (Prevision Of Precipitation)
least_rain_cities_df = cities_df.sort_values(by="meteo",ascending=True)

# Converting lat & lon to float >> Should be done on cities_df creation!
cities_df["lat"] = cities_df["lat"].astype(float)
cities_df["lon"] = cities_df["lon"].astype(float)

# Saving sorted cities meteo prevision to a cities.csv file
least_rain_cities_df.to_csv("cities.csv")
```

```python
least_rain_cities_df['marker_size'] = 7
least_rain_cities_df['proba_of_rain'] = least_rain_cities_df['meteo']
# Displaying a map of cities
fig = px.scatter_mapbox(least_rain_cities_df.head(),
                        lat="lat",
                        lon='lon',
                        color='proba_of_rain',
                        size='marker_size',
                        hover_name="name",
                        hover_data="proba_of_rain",
                        mapbox_style="open-street-map",
                        color_continuous_scale=px.colors.carto.Agsunset_r,
                        zoom=5,
                        height=600,
                    )
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.write_image("src/top5_cities_by_meteo.png")
fig.show()
```
